```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from pylab import *
x = np.linspace(0, 5, 10)
y = x ** 2

figure()
plot(x, y, 'r')
xlabel('x')
ylabel('y')
title('MATLAB-like API')
show()
```

```python
from urllib.request import urlopen
import pandas as pd
diamonds = pd.read_csv(urlopen('http://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/csv/ggplot2/diamonds.csv'))
titanic = pd.read_csv(urlopen('http://www.columbia.edu/~cjd11/charles_dimaggio/DIRE/resources/R/titanic.csv'))
```

```python
fig, axes = plt.subplots(2,1, figsize=(12,6))
axes[0].hist(diamonds.depth, bins=np.linspace(50, 70, 200))
axes[0].set_xlabel("Depth")
axes[0].set_xlim([55, 70])

axes[1].hist(diamonds.depth, bins=np.linspace(50, 70, 40))
axes[1].set_xlabel("Depth")
axes[1].set_xlim([55, 70])
```

```python
x = diamonds.carat[diamonds.carat < 2]
y = diamonds.price[diamonds.carat < 2]
fig, axes = plt.subplots()
axes.plot(x, y, 'o', mec='none', alpha=.05)

#приблизим полиномом второго порядка
params = np.polyfit(x, y, 2)
xp = np.linspace(x.min(), 2, 20)
yp = np.polyval(params, xp)
axes.plot(xp, yp, 'k')

axes.set_xlabel("Carat")
axes.set_ylabel("Price")
axes.axis('tight')
axes.set_xlim(0, 2)
```

```python
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'times new roman'})
classes = titanic.groupby(['pclass']).size()
fig, axes = plt.subplots()
axes.set_aspect(1)
axes.pie(classes, labels=classes.index.values, colors=plt.rcParams['axes.color_cycle'])
axes.set_title("Passenger Class on the Titanic")
```

```python
tclass = titanic.groupby(['pclass', 'survived']).size().unstack()
print(tclass)

fig, axes = plt.subplots(1, 2, figsize=(10, 6))
axes[0].bar([0, 1, 2], tclass[0], color='maroon', label='Died')
axes[0].bar([0, 1, 2], tclass[1], bottom=tclass[0], color='midnightblue', label='Survived')
axes[0].set_xticks([0.5, 1.5, 2.5])
axes[0].set_xticklabels(['1st Class', '2nd Class', '3rd Class'], rotation=0)
axes[0].set_ylabel("Number")
axes[0].set_xlabel("")
axes[0].legend(loc='upper left')

#нормализуем классы с помощью транспонирования, нормализации столбцов и обратного транспонирования
tclass = (1. * tclass.T / tclass.T.sum()).T

axes[1].bar([0, 1, 2], tclass[0], color='maroon', label='Died')
axes[1].bar([0, 1, 2], tclass[1], bottom=tclass[0], color='midnightblue', label='Survived')
axes[1].set_xticks([0.5, 1.5, 2.5])
axes[1].set_xticklabels(['1st Class', '2nd Class', '3rd Class'], rotation='horizontal')
axes[1].set_ylabel("Fraction")
axes[1].set_xlabel("")
fig.tight_layout()
```

```python
matplotlib.style.use('ggplot')
tclass = titanic.groupby(['sex', 'survived']).size().unstack()
fig, axes = plt.subplots(1, 2, figsize=(10,6))
axes[0].bar([0, 1], tclass[0], color='maroon', label='Died')
axes[0].bar([0, 1], tclass[1], bottom=tclass[0], color='midnightblue', label='Survived')
axes[0].set_xticks([0.5, 1.5])
axes[0].set_xticklabels(['Female', 'Male'], rotation=0)
axes[0].set_ylabel("Number")
axes[0].set_xlabel("")
axes[0].legend(loc='upper left')

tclass = (1. * tclass.T / tclass.T.sum()).T
axes[1].bar([0, 1], tclass[0], color='maroon', label='Died')
axes[1].bar([0, 1], tclass[1], bottom=tclass[0], color='midnightblue', label='Survived')
axes[1].set_xticks([0.5, 1.5])
axes[1].set_xticklabels(['Female', 'Male'], rotation=0)
axes[1].set_ylabel("Number")
axes[1].set_xlabel("")
```

```python
from mpld3 import display

fig, axes = plt.subplots()
np.random.seed(0)
axes.plot(np.random.normal(size=100),
          np.random.normal(size=100),
          'or', ms=10, alpha=0.3)
axes.plot(np.random.normal(size=100),
          np.random.normal(size=100),
          'ob', ms=20, alpha=0.1)

axes.set_xlabel('Ось X')
axes.set_ylabel('Ось y')
axes.set_title('Matplotlib, визуализированный в D3!', size=14)
axes.grid(color='lightgray', alpha=0.7)

display(fig)
```
